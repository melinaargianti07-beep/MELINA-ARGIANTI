```python
import matplotlib.pyplot as plt

g = 9.8
h = 100
t = list(range(11)) #0 sampai 10 detik
s = [0.5 * g * ti**2 for ti in t]
v = [g * ti for ti in t]
a = [g] * len(t)


# Plot
plt.plot(t, s, marker='o', label='Posisi (m)')
plt.plot(t, v, marker='s', label='Kecepatan (m/s)')
plt.plot(t, a, marker='^', label='Percepatan (m/s²)')
plt.title("Gerak Jatuh Bebas (Kasus 1)")
plt.xlabel("Waktu (detik)")
plt.ylabel("Nilai")
plt.grid(True)
plt.legend()
plt.show()

# Nilai maksimum
print("Kasus 1:")
print(f"Posisi maksimum: {max(s):.2f} m")
print(f"Kecepatan maksimum: {max(v):.2f} m/s")

```
